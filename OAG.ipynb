```python
'''登陆JobBin导出数据链接'''
#直接获取
name = 'jwang285@mail.tsinghua.edu.cn'
password = 'Z7PXgX.v'
url='https://analytics.oag.com/analyser-client/home'

browser = webdriver.Chrome()
browser.get(url)
time.sleep(5)

# 输入用户名
input_name=browser.find_element(By.XPATH,'//*[@id="login-wrapper"]/div/div/div[1]/input')
input_name.clear()
input_name.send_keys(name)

# 输入密码
input_pass=browser.find_element(By.XPATH,'//*[@id="login-wrapper"]/div/div/div[2]/input')
input_pass.clear()
input_pass.send_keys(password)

# 点击登陆
login_but=browser.find_element(By.XPATH,'//*[@id="login-wrapper"]/div/div/div[3]/button')
login_but.click()
time.sleep(2)

# 进入Schedulers Analyser界面
traffic_analy=browser.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div[2]/div[1]/div[2]/h2/a')
traffic_analy.click()
time.sleep(2)

# 下滑页面
browser.execute_script("window.scrollTo(0,document.body.scrollHeight);")
time.sleep(10)

# 点击Open Jobbin
open_jobbin=browser.find_element(By.XPATH,'//*[@class="button noop"]')
browser.execute_script("arguments[0].click();", open_jobbin)
time.sleep(15)

# 下滑页面
browser.execute_script("window.scrollTo(0,document.body.scrollHeight);")
time.sleep(3)
    
# 先跳转至最后一页，再回到第一页
last_page = browser.find_element(By.XPATH, '//span[@class="record-last margin_right_5px"]') 
browser.execute_script("arguments[0].click();", last_page)
time.sleep(5)

# 下滑页面
browser.execute_script("window.scrollTo(0,document.body.scrollHeight);")
time.sleep(3)

# 需要点击一次最后一页，防止页面自动跳转回首页
last_page = browser.find_element(By.XPATH, '//span[@class="record-last margin_right_5px"]') 
browser.execute_script("arguments[0].click();", last_page)
time.sleep(5)

first_page = False
download_list = []

while first_page==False:
    soup = BeautifulSoup(browser.page_source, 'html.parser')
   
    # 判断页面是否自动跳转
    if soup.find('span', class_="left record-count-label").text[:6]=='1 - 10':
        first_page = True

    # 获取存储OAG数据的tr标签
    for tr in soup.find('tbody').find_all('tr'):
        tds = tr('td')
        if tds[7].text.strip()[:12] == '06 Sept 2023':
            href = tds[9].a['href']
            print(href)
            download_list.append(href)   
    
    # 向前翻页
    previous_page = browser.find_element(By.XPATH, '//span[@class="record-previous margin_right_5px"]') 
    browser.execute_script("arguments[0].click();", previous_page)
    time.sleep(1)
```

```python
'''下载文件并保存'''
# 设置headers
headers={'Connection':'close'}
for href in download_list:
    content = requests.get(href, headers=headers, timeout=30, verify=False).content
    # 写入文件
    with open("OAG/" + href[-21:], "wb") as code:
        code.write(content)
    time.sleep(random.randint(1, 3))
```

```python
import pandas as pd
import glob
import os

# 设置路径
path = "'/OAG"

city_airport['Frequency'] = None
city_airport['Time'] = None

# 使用glob读取多个文本文件 这里选取的是excel的后缀
for file in glob.glob(os.path.join(path,"*.xlsx")):
    print(file)
    # 修改数据格式
    data = pd.read_excel(file)
    # 根据格式筛选数据
    try:
        data = data.iloc[21:-2, :]
    except:
        data = data.iloc[23:-2, :]
    data.columns = ['Frequency', 'Time series']
    # 提取Airport code
    airport_code = file[-21:-18]
    # 提取数据
    time = data.iloc[0, 1]
    frequency = data['Frequency'].sum()
    print(frequency, time)

    city_airport.loc[city_airport['Airport Code']==airport_code, 'Frequency'] = frequency
    city_airport.loc[city_airport['Airport Code']==airport_code, 'Time'] = time
# 保存文件
city_airport.to_excel('city_airport.xlsx')
```

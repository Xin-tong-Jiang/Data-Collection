```python
import os
import glob
import pandas as pd
import selenium 
from  selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
data = pd.read_excel('C:/Users/蒋欣彤/Desktop/文件3.xlsx')
data.head()
```

```python
city_list = []
for i in range(len(data)):
    city = data.iloc[i, 6] + ', ' + data.iloc[i, 3]
    city_list.append(city)
    
city_list
```

```python
browser = webdriver.Chrome()
browser.get('https://osiris.bvdinfo.com/version-2022825/home.serv?product=OsirisNeo&&SetLanguage=en&')
time.sleep(3)
```

```python
##########登陆##########
# 输入用户名：输入你的用户名
browser.find_element(By.XPATH,'//*[@id="username"]').send_keys('')
# 输入密码：输入你的密码
browser.find_element(By.XPATH,'//*[@id="password"]').send_keys('')
# 点击登陆
browser.find_element(By.XPATH,'//*[@id="organic-div"]/form/div[3]/button').click()
time.sleep(5)
```

```python
##########创建人才库报告##########
browser.find_element(By.XPATH,'/html/body/div[4]/div[7]/main/section[1]/div/a[1]').click()#菜单
time.sleep(3)
```

```python
##########设置筛选条件##########
for city in city_list:
    try:
        # 点击地点+
        browser.find_element(By.XPATH,'/html/body/div[4]/div[7]/div/div/div/div/div/div/search-filters-panel-content/div/aside/div[1]/div[2]/div/div[1]/button').click()
        time.sleep(3)
        # 输入地点
        location = browser.find_element(By.XPATH,'/html/body/div[4]/div[7]/div/div/div/div/div/div/search-filters-panel-content/div/aside/div[1]/div[2]/div/div[2]/section/div/div/div/span/div/div[2]/div/input').send_keys(city)
        time.sleep(3)
        # 选择第一个作为结果
        browser.find_element(By.XPATH,'/html/body/div[4]/div[7]/div/div/div/div/div/div/search-filters-panel-content/div/aside/div[1]/div[2]/div/div[2]/section/div/div/div/span/div/ul/div/div/li[1]').click()
        time.sleep(3)
        # 点击apply
        apply = browser.find_element(By.XPATH,'/html/body/div[4]/div[7]/div[1]/div/div/div/div/div/search-filters-panel-content/div/aside/div[4]/div/button')
        browser.execute_script("arguments[0].click();", apply)
        time.sleep(10)

        # 点击所在地点
        browser.find_element(By.XPATH, '/html/body/div[4]/div[7]/div[1]/section/div[1]/nav/a[2]').click()
        time.sleep(10)

        # 点击人才主要集中地点：查看
        browser.find_element(By.XPATH, '/html/body/div[4]/div[7]/div[1]/section/main/div[1]/div/div[1]/div/div[2]/div[1]/div/button/div').click()
        time.sleep(5)
        # 点击市/区
        browser.find_element(By.XPATH,'/html/body/div[4]/div[7]/div[1]/section/main/div[1]/div/div[1]/div/div[2]/div[1]/div/div/div/ul/li[1]/div').click()
        time.sleep(8)
        
        try:
            # 点击选择地点
            browser.find_element(By.XPATH, '/html/body/div[4]/div[7]/div[1]/section/main/div[2]/div/div[2]/div[1]/div[2]/button').click()
            time.sleep(5)
            # 选择第一个
            browser.find_element(By.XPATH, '/html/body/div[4]/div[7]/div[1]/section/main/div[2]/div/div[2]/div[1]/div[2]/div/div/ul/li').click()
            time.sleep(8)
            
        except:
            print(city, 'confirmed!')

        # 点击所选地点招募和流失人才的地点：查看
        browser.find_element(By.XPATH,'/html/body/div[4]/div[7]/div[1]/section/main/div[2]/div/div[2]/div[2]/div/button').click()
        time.sleep(5)
        # 点击国家/地区
        browser.find_element(By.XPATH,'/html/body/div[4]/div[7]/div[1]/section/main/div[2]/div/div[2]/div[2]/div/div/div/ul/li[4]/div').click()
        time.sleep(8)

        # 点击省略号导出
        down_load = browser.find_elements(By.XPATH, '/html/body/div[4]/div[7]/div[1]/section/main/div[2]/div/div[1]/div/div[2]/div/button/li-icon')
        down_load[0].click()
        time.sleep(3)
        # 点击excel
        browser.find_elements(By.XPATH, '/html/body/div[4]/div[7]/div[1]/section/main/div[2]/div/div[1]/div/div[2]/div/div/div/ul/li[2]/div/li-icon')[0].click()
        time.sleep(12)
        
        print(city, 'succeed!')

        # 清除搜索结果
        browser.find_element(By.XPATH, '/html/body/div[4]/div[7]/div[1]/div/div/div/div/div/search-filters-panel-content/div/aside/div[1]/div[2]/div/div[1]/button[2]').click()
        time.sleep(3)
        
    except:
        print(city, 'fail!')
        # 清除搜索结果
        browser.find_element(By.XPATH, '/html/body/div[4]/div[7]/div[1]/div/div/div/div/div/search-filters-panel-content/div/aside/div[1]/div[2]/div/div[1]/button[2]').click()
        time.sleep(3)
        continue
```

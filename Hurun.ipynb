```python
import bs4
from bs4 import BeautifulSoup
from lxml import etree
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import random

# 设置头文件
headers = {
    'user-agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36',
    'Cookie':'Hm_lvt_2b09927a5895e3946dc6de8526befc81=1693446797; Hm_lpvt_2b09927a5895e3946dc6de8526befc81=1693446797; __utma=245691549.570358353.1693446797.1693446797.1693446797.1; __utmc=245691549; __utmz=245691549.1693446797.1.1.utmcsr=(direct)|utmccn=(direct)|utmcmd=(none)'
}

def get_company(url):
    # 初始化浏览器自动化工具
    driver = webdriver.Chrome()
    # 获取页面
    driver.get(url)
    # 设置休眠时间
    time.sleep(4)
    
    # 创建数据表存储结果
    data = pd.DataFrame(columns=['Rank', 'Wealth', 'Enterprise Information', 'Leaders / Founders', 'Industry'])
    # 设置页面范围
    page_num = 1

    while page_num < 69:
        # 解析页面
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        # 获取存储学校数据的tr标签
        for tr in soup.find('tbody').children:
            if isinstance(tr, bs4.element.Tag):
                # 学校名称存储在标签<a>中，其余存储在各个td中
                tds = tr('td')
                RANK = tds[0].p('em')[1].text.strip()
                WEALTH = tds[1].p.text.strip()
                ENTERPRISE = tds[2].div.div.div.div.div.div.span.text.strip()
                LEADERS = tds[3].div.p.text.strip().replace('\xa0','')
                INDUSTRY = tds[4].div.p.text.strip()
                # 将结果写入数据表
                value = [RANK, WEALTH, ENTERPRISE, LEADERS, INDUSTRY]
                data.loc[len(data.index)] = value
                # 打印结果
                print(value)
                
        # 获取翻页按钮
        next_page = driver.find_element(By.XPATH,'(//li[@class="page-item page-next"])')
        driver.execute_script("arguments[0].click();", next_page)
                                   
        time.sleep(random.randint(1, 3))
        
        page_num += 1
        
    data.to_excel('Hurun.xlsx', index=False)

url = 'https://www.hurun.net/en-US/Rank/HsRankDetails?pagetype=unicorn'
data = get_company(url)
```

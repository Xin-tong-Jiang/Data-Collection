#### Speedtest

```python
import bs4
from bs4 import BeautifulSoup
from lxml import etree
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import random

def get_Speedtest(url):
    # 创建数据表存储结果
    data = pd.DataFrame(columns=['Rank','Rank Change', 'City', 'Country', 'Mobile(Mbps)'])
    
    # 初始化浏览器自动化工具
    driver = webdriver.Chrome()    
    driver.get(url)    # 获取页面
    time.sleep(4)    # 设置休眠时间
    
    switch_button = driver.find_element(By.XPATH,'(//div[@class="toggle__switch"])')
    switch_button.click()
    show_button = driver.find_element(By.XPATH,'(//button[@class="btn"])')
    show_button.click()
    time.sleep(4)    # 设置休眠时间
    
    # 解析页面
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    tbody = soup.find_all('tbody')[3]    # 所需数据在第四个tbody下
    trs = tbody.find_all('tr', class_="data-result results")    # 找到全部行
    # 获取全部tr标签
    for tr in trs:
        tds = tr.find_all('td')
        value = [tds[0].text, tds[1].span.text, tds[2].a.text.strip(), tds[2].span.text, tds[3].text]
        # 将结果写入数据表
        data.loc[len(data.index)] = value
        # 打印结果
        print(value)
        
    #     # 将结果写入excel       
    data.to_excel('Speedtest.xlsx', index=False)

url = 'https://www.speedtest.net/global-index'
get_Speedtest(url)
```

```python
import bs4
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
import time
import random

# 设置头文件
headers = {
    'user-agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36',
    'Cookie':'_ga=GA1.1.384274012.1692500719; JSESSIONID=CC3B53129A2D284ED57C289440062E32.savvycms; CFID=13420391; CFTOKEN=4718b3c4a5ad2815-B170C06E-0A40-7ECB-15D28BE35A550469; _ga_XDQGWKHM3H=GS1.1.1692944338.2.1.1692945176.59.0.0'
}

def get_people(url):
    data = pd.DataFrame(columns=['TIME', 'NAME', 'LINK'])
    
    # 获取获奖人数据
    html = requests.get(url, headers=headers, verify=False).text
    soup = BeautifulSoup(html, 'html.parser')

    # 获取存储学校数据的tr标签
    awl = soup.find_all('ul', class_='award-winners-list')
    awll = soup.find('ul', class_='award-winners-list last')
    awl_list = list(awl) + list(awll)

    for aw in awl_list[:3]:
        for li in aw.find_all('li'):
            if isinstance(li, bs4.element.Tag):
                # 获取企业数据及页面链接
                TIME = li.span.text.strip()[1:-1]
                for a in li.find_all('a'):
                    NAME = a.text.strip()
                    LINK = 'https://amturing.acm.org' + a['href']
                    value = [TIME, NAME, LINK]
                    data.loc[len(data.index)] = value
                    # 打印结果
                    print(value)
                    
    data.to_excel('AM Turing(temp).xlsx', index=False)
    return data

url = 'https://amturing.acm.org/byyear.cfm'
data = get_people(url)
# result = get_personal_info(data)
```
